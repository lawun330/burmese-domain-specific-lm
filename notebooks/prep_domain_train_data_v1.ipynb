## Setup

````python
from pathlib import Path
import pandas as pd
````

````python
ROOT = Path("..").resolve()

TRAIN_DIR = ROOT / "data" / "train" / "domain-specific"
LOG_DIR = TRAIN_DIR / "logs"
LOG_DIR.mkdir(parents=True, exist_ok=True)

CLEAN_TEXT = ROOT / "clean_text.py"

SYL_NORM = ROOT / "syl-normalizer" / "syl_normalizer.py"
SYL_DICT = ROOT / "syl-normalizer" / "final_syl_dictionary_13Feb2024.sorted.txt"

OPPA_WORD = ROOT / "oppaword" / "oppa_word.py"
OPPA_DICT = ROOT / "oppaword" / "myg2p_mypos.dict"
````

## Preprocess All Datasets: News Domain

### Remove Punctuations and Word Tags

````python
!python {CLEAN_TEXT} -i {TRAIN_DIR}/train_news.raw -o {TRAIN_DIR}/train_news.cleaned.state1
````

### Syllable Normalization

````python
!python {SYL_NORM} \
    --dictionary {SYL_DICT} \
    --frequency 2 \
    --input {TRAIN_DIR}/train_news.cleaned.state1 \
    --output {TRAIN_DIR}/train_news.cleaned.state2 \
    --log {TRAIN_DIR}/logs/train_news.cleaned.state2.log \
    --error-output ../data/train/logs/train_news.cleaned.state2.errors \
    --fuzzy-distance 0
````

### Word Segmentation (OppaWord)

````python
!python {OPPA_WORD} \
  --input {TRAIN_DIR}/train_news.cleaned.state2 \
  --dict {OPPA_DICT} \
  --output {TRAIN_DIR}/train_news.cleaned.state3
````
